# Simple numerical intergration for motion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Attempt 1, throwing the stone upwards
a = -1           #Acceleration (-bc it tries to limit the flight of the stone)
v = 5            #We throw the stone upwards so it's positive
x = 0            # We throw it from the surface (0)
dt = 0.01        # Our step. Numerical intergration step.  Is supposed to be small
nsteps = 1000
history = np.zeros(nsteps)  # This we will plot that eventually 
for i in range(0, nsteps):
    v = v + a*dt
    x = max(x + v*dt,0)
    history[i] = x

plt.plot(history);
```
