```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
x = np.arange(0,n,1)
p = np.zeros(n)
p[0] = 1
p[1] = 1
not_sharing = 1
for i in range (0,n):
    not_sharing *= (1 - (i/365))
    p[i] = 1 - not_sharing
plt.plot(x,p);
```

# Alternative (cleaner) Solution

```python
maxn = 80
p = np.zeros(maxn)
p[0] = 1
p[1] = 1
for i in range (2, maxn):
    p[i] = p[i-1] *(1-(i-1)/365)
plt.plot(1 - p)
plt.xlabel ('N people in the group')
plt.ylabel ('P of no shared BD');
```
