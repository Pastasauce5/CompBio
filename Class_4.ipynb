```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nstep = 50
figure = plt.figure()
x = .66
a = 0
for r in np.arange(2.8,4,.0001):
    a = a + 1
    for i in range(nstep + a % 50):
        x = x*r*(1-x)
    plt.plot(r,x,'k.', ms = 1)
#It's like marrying your mom, they shold be 2 different people
```

```python
nsteps = 100 
x = np.zeros(nsteps)
x[0] = .6
r = 3.9
plt.figure(figsize=(12,5))
A = ([.60,.601])
for x[0] in A:
    for i in range (1,nsteps):
        x[i] = x[i-1]*r*(1-x[i-1])
    plt.plot(x);
    plt.ylim(0,1)
```

```python
nsteps = 100
ntraces = 10
r= 4 
wiggle = 0.0001
x = np.zeros(shape = (nsteps,ntraces)) #Creates a matrix of zeros
for itrace in range(ntraces):
    x[0,itrace] = 0.6 + np.random.randn(1)*wiggle
    for istep in range(1,nsteps):
        x[istep,itrace] = r*x[istep - 1, itrace]*(1 - x[istep-1, itrace])

plt.xlabel('Time')
plt.ylabel('All trajectories')
plt.plot(x)
plt.plot(np.std(x,1),'k');
```
