```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
n_animals = 100
n_steps = 100
speed = .01
x = np.ones(n_animals)*.5
y = np.ones(n_animals)*.5

for i in range (n_steps):
    x += np.random.randn(n_animals)*speed
    y += np.random.randn(n_animals)*speed
    
fig = plt.figure(figsize = (3,3))
plt.plot(x,y, 'k.')
plt.xlim(0,1)
plt.ylim(0,1);
```

```python
%matplotlib notebook

n_animals = 100
n_steps = 100
speed = .05
x = np.ones(n_animals)*.5
y = np.ones(n_animals)*.5

fig = plt.figure(figsize = (5,5))
plt.xlim(0,1)
plt.ylim(0,1);

for i in range (n_steps):
    x += np.random.randn(n_animals)*speed
    y += np.random.randn(n_animals)*speed
    x[x<0] = 0;x[x>1] = 1
    y[y<0] = 0;y[y>1] = 1
    plt.cla()
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.plot(x,y, 'k.', markersize = 5)
    fig.canvas.draw()
    time.sleep(0.1)
    

```

## Better ways to limit values in a numpy vector
The power of numpy is that is we can write a one-liner. it will be really fast. So we should try to avoid for loops, if possible

For a playground situation let's create a numpy array with only one negative element, and then let's find a way to set this element equal to 0.  So like limiting from below in our previous model

```python
a = np.array([1,2,3,4,-7,5])
#option one
for i in range(len(a)):
    if a[i]<0: a[i] = 0
print (a)

#Option 2, boolena indexing
a = np.array([1,2,3,4,-7,5])
print('a<0', a<0) 
print('a[a<0]:',a[a<0]) #Logical indexing
a[a<0] = 0
print(a)

#Option 3, vectorized formula
a = np.array([1,2,3,4,-7,5])
a = np.maximum(a,0)
print(a)
```
