Predator Prey Models

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100000
x = np.zeros(nsteps) #number of bunnies
y = np.zeros(nsteps) #number of foxes
#Delta x = ax(birth rate)-bxy(death rate, both by natural causes and by predators)
#Delta y = cxy(constant represents amount of energy the predators get from prey and how it translates to birthrate) + 
#dy(natural death rate)
a = 0.6
b = 1.2
c = .5
d = 1
dt = 0.001
possible_ds = [.25, .5, .75, 1]
for d in possible_ds:
    x[0] = 1
    y[0] = 1

    for i in range(1,nsteps):
        x[i] = x[i-1] + dt*(a*x[i-1] - b*x[i-1]*y[i-1])
        y[i] = y[i-1] + dt*(c*x[i-1]*y[i-1] - d*y[i-1])

    fig = plt.figure()
    plt.plot(x)
    plt.plot(y)
```
