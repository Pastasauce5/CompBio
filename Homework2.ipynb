```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 50 #Number of rows in our matrix
ntraces = 10 #NUmber of columns in our matrix
x = np.zeros(shape=(nsteps,ntraces)) #Placeholder for trajectories we will generate
wiggle = 0.001 #The standard deviation
r = 5 
for i in range(r):
    for itrace in range(ntraces): #This runs the outside loop 20 times
        x[0,itrace] = 0.6 + np.random.randn(1) * wiggle #This creates the randomness in each row
        for istep in range (1,nsteps): # Another loop to address each point in a row in the matrix
    # while adjusting the point for the randomness created in the outside loop.
            x[istep,itrace]= r*x[istep-1,itrace] * (1-x[istep-1,itrace])  
    plt.figure()
    plt.plot(x[i])
    plt.xlabel('Time');
    plt.ylabel('All Possibilities');
```
