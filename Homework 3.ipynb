```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 20000
a = 2  #limited growth rate of bunnies 
b = 1.2
c = 0.5
d = 1
e = 0.3 # energy that wolves will get from foxes 
f = 0.8 #death rate of wolves
dt = 0.001

possible_ds = [0.5, 1, 1.5, 2]

x = np.zeros(nsteps)
y = np.zeros(nsteps)
z = np.zeros(nsteps)

fig1 = plt.figure(figsize=(16,5));
fig2 = plt.figure(figsize=(5,5));
ax2 = plt.axes()

for itrial in range(len(possible_ds)):
    d = possible_ds[itrial]
    x[0] = 1
    y[0] = 1
    z[0] = 1

    for t in range(1,nsteps):
        x[t] = x[t-1] + dt*(a*x[t-1] - b*x[t-1]*y[t-1])
        y[t] = y[t-1] + dt*(c*x[t-1]*y[t-1] - d*y[t-1] - e*y[t-1]) #Added death rate due to wolves
        z[t] = z[t-1] + dt*(e*x[t-1]*z[t-1] - f*z[t-1]) #Wolves

    ax1 = fig1.add_subplot(1,len(possible_ds),itrial+1)
    ax1.plot(x, label='Bunnies');
    ax1.plot(y, label='Foxes');
    ax1.plot(z, label = 'Wolves')
    ax1.legend(loc='upper right');

    ax2.plot(x,y, label=d);
    plt.xlabel('Bunnies');
    plt.ylabel('Foxes');
 
ax2.legend(loc='upper right');
```

For this I simply added a new factor into the system (wolves) with their own predation rate and their own birth rate.  Then the value that I changed was the death rate of the foxes.
