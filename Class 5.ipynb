```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100
ntraces = 10
r= 4 
wiggle = 0.0001
diff = [nsteps,2]
x = np.zeros(shape = (nsteps,ntraces)) #Creates a matrix of zeros
for i in range(r):
    for itrace in range(ntraces):
        x[0,itrace] = 0.6 + np.random.randn(1)*wiggle
        for istep in range(1,nsteps):
            x[istep,itrace] = r*x[istep - 1, itrace]*(1 - x[istep-1, itrace])

plt.xlabel('Time')
plt.ylabel('All trajectories')
plt.plot(x)
plt.plot(np.std(x,1),'k');

    
```

```python
nsteps = 100
ntrials = 100
x = np.zeros(shape = (nsteps, 2))
wiggle = 0.001
diff = np.zeros(shape = (nsteps, ntrials))
for r in np.arange(3.6,4,0.1):
    for itrial in range(ntrials):
        starting_point = np.random.uniform(high = 1-wiggle, low = 0) + np.array([0,wiggle])
        for ipair in range(2):
            x[0,ipair] = starting_point[ipair]
            for istep in range(1,nsteps):
                x[istep, ipair] = r*x[istep - 1, ipair]*(1 - x[istep-1, ipair])
        diff[:, itrial] = np.abs(x[:,1]-x[:,0])
    plt.plot(np.mean(diff, axis = 1))
  
    
```
